```python
########## Let's Run An Online Analysis ##########

## Let's say that the true theta is 0.75
θ = 0.75

## Initialize the agent and plot the prior
jake = BetaBinomialAgent(1, 1)
jake.plot()

for _ in range(200):
    flip = np.random.binomial(n=1, p=θ, size=1)
    jake.update(flip)
```

```python
## Plot the posterior
jake.plot()
```
